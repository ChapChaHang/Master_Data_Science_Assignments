#### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

#### Key difference between Pandas series and Dataframe:
- Series is a one-dimensional labeled array that can hold data of any type, whereas a DataFrame is a 2-dimensional labeled data structure with columns of potentially different types.

- Series represents a single column of data, while a DataFrame represents a collection of columns (i.e., a table).

- Series has only one axis (i.e., a row index), while a DataFrame has two axes (i.e., row and column indices).

- Series does not have column labels, while a DataFrame has column labels that can be used to access individual columns of data.

- Series can be created from a list, tuple, array, or dictionary, while a DataFrame can be created from multiple Series objects, a list of lists, a dictionary of lists, or a structured numpy array.

- Operations on a Series are performed on a single column of data, while operations on a DataFrame can be performed on multiple columns or rows of data.

- A Series can be used as a building block for a DataFrame (i.e., a DataFrame can be constructed by combining multiple Series objects), while a DataFrame cannot be used as a building block for a Series.

#### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

#### common funtions for data manipulation of Pandas DataFrame are as follows:

```python
df1 = pd.read_csv('services.csv')
```

#### head () = Returns the first n rows of the DataFrame,

```python
df1.head(4)
```

#### tail() = Returns the last n rows of the DataFrame,

```python
df1.tail(4)
```

#### info() = Provides information about the DataFrame, including the number of rows, number of columns, column names, data types, and memory usage.

```python
df1.info()
```

#### describe() = Generates descriptive statistics of the DataFrame

```python
df1.describe()
```

#### sort_values() = Sorts the DataFrame by one or more columns

```python
df1.sort_values('location_id',ascending=False)
```

####  groupby() = Groups the DataFrame by one or more columns and performs an aggregation function on each group

```python
df1.groupby('status').size()
```

```python
df1['status'].unique()
```

#### fillna() = Fills missing values in the DataFrame with a specified value or method

```python
df1.isna().sum()
```

```python
df2 = df1.fillna('filled')
```

```python
df2.head()
```

```python
df2.isna().sum()
```

#### df.loc() =  This method is used to select rows and columns from a DataFrame based on their labels. 

```python
df2.columns
```

```python
df2.loc[0:5,['id', 'location_id', 'program_id']]
```

#### df.iloc() = This method is used to select rows and columns from a DataFrame based on their integer positions.

```python
df2.iloc[0:5,[0,1,2]]
```
