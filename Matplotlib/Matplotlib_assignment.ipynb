### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

- import numpy as np
- np.random.seed(3)
- x = 3 + np.random.normal(0, 2, 50)
- y = 3 + np.random.normal(0, 2, len(x))

```python
import matplotlib.pyplot as plt
```

#### Scatter Plot: A scatter plot is used to visualize the relationship between two variables. It displays the data points as a collection of points.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y,c='#000080',alpha=0.7,marker='.')
plt.xlabel('Random numer of X-axis')
plt.ylabel('Random number of y-axis')
plt.title("Example of Scatter Plot using Matplotlib")
plt.grid()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])  
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

#### The subplot() function in Matplotlib is used to create multiple plots in the same figure. It allows us to create a grid of plots, where each plot is a separate subplot within the same figure.

```python
fig,test = plt.subplots(nrows=2,ncols=2)
fig.suptitle('Example of SubPlots')
test[0,0].plot(x1,y1)
test[0,0].set_title("First Line Plot")
test[0,1].plot(x2,y2)
test[0,1].set_title("Second Line Plot")
test[1,0].plot(x3,y3)
test[1,0].set_title("Third Line Plot")
test[1,1].plot(x4,y4)
test[1,1].set_title("Fourth Line Plot")
fig.tight_layout()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

#### A bar plot is a type of plot in which rectangular bars are used to represent the data. Bar plots are typically used to compare different categories of data and to visualize the distribution of a variable across different categories. The height or length of each rectangular bar represents the frequency of an event. 

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(6,6))
plt.bar(company,profit)
plt.title("Bar Plot")
plt.ylabel('Profit')
plt.xlabel('Company')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.barh(company,profit)
plt.title(" Horizontal Bar Plot")
plt.ylabel('Company')
plt.xlabel('Profit')
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

#### A box plot is  a type of plot that displays the distribution of a dataset through five key summary statistics: the minimum value, the maximum value, the median (or 50th percentile), the first quartile (or 25th percentile), and the third quartile (or 75th percentile). Box plots are used to display the distribution of a dataset in a way that is easily understandable and allows for easy comparison between groups. They are particularly useful when dealing with large datasets, as they can provide a visual summary of the data in a compact form.Box plots can also be used to identify potential outliers, as well as to detect differences or similarities between groups of data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1,box2]
```

```python
plt.boxplot(data,patch_artist=True)
plt.show()
```
