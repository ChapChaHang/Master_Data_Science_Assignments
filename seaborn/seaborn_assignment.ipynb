### 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

- Scatter plot: A scatter plot is used to visualize the relationship between two continuous variables. 

- Histogram: A histogram is used to show the distribution of a continuous variable. 

- Bar plot: A bar plot is used to compare the values of categorical data.

- Box plot: A box plot is used to show the distribution of a continuous variable across different categories. 

- Heatmap: A heatmap is used to visualize the relationship between two categorical variables or one categorical variable and one continuous variable.

### 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(data=fmri, x=fmri.timepoint,y=fmri.signal,hue='event',style='region')
```
